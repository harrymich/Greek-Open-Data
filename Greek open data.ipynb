```python
import numpy as np
import pandas as pd
import shapefile as shp
import matplotlib.pyplot as plt
import contextily as ctx
import geopandas as gpd
import seaborn as sns
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

# Αιολικό δυναμικό

```python
data =  gpd.read_file('aiolikos_charths.shp')
data = data.to_crs(epsg=3857)
```

```python
# plot the data using geopandas .plot() method
fig, ax = plt.subplots(figsize = (10,10))
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)
data.plot(column='WV',ax=ax,legend=True,markersize=0.2,cmap='GnBu',vmax=15,legend_kwds={'label': "Wind speed at 40m elevation (m/s)"},cax=cax)
ctx.add_basemap(ax, source=ctx.sources.ST_TERRAIN_BACKGROUND)
plt.show()
```

# Θέρμανση (Βαθμοημέρες θέρμανσης)

```python
hdd =  gpd.read_file('vathmohmeres_thermanshs.shp')
hdd = hdd.to_crs(epsg=3857)
```

```python
fig, ax = plt.subplots(figsize = (10,10))
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)
hdd.plot(column='TOT18',ax=ax,legend=True,cmap='coolwarm_r',legend_kwds={'label': "Heating Degree-Days"},cax=cax)
ctx.add_basemap(ax, source=ctx.sources.ST_TERRAIN_BACKGROUND)
```
